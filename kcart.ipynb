```python
def clean_numeric_column(series):
    """Ensures all numeric columns remain floats with proper precision."""
    
    # Convert entire column to string and clean unwanted characters
    cleaned = (
        series.astype(str)
        .str.strip()
        .str.replace(r"[\'‘’\"“”`]", "", regex=True)  # Remove quotes
        .str.replace(",", "", regex=True)  # Remove thousands separator (13,853 → 13853)
        .str.replace(r"[^\d.\-]", "", regex=True)  # Keep only numbers, dots, and minus
        .str.replace(r"\.{2,}", ".", regex=True)  # Remove multiple dots
        .str.replace(r"(?<!\d)\.(?!\d)", "", regex=True)  # Remove lone dots
        .str.rstrip('.')  # Remove trailing dots
    )

    # Convert to float and handle errors
    cleaned = pd.to_numeric(cleaned, errors='coerce')  # NaN for errors
    cleaned = cleaned.apply(lambda x: round(x, 6) if pd.notnull(x) else np.nan)

    return cleaned

df_Limestone = dfs["Limestone pile"]

print(df_Limestone.dtypes)

date_column_index = df_Limestone.columns.get_loc('DATE')
columns_to_modify = df_Limestone.columns[date_column_index + 1:]
new_columns = ["Limestone " + str(col) for col in columns_to_modify]
# Create a dictionary to map old column names to new column names
column_mapping = {old_col: new_col for old_col, new_col in zip(columns_to_modify, new_columns)}
df_Limestone.rename(columns=column_mapping, inplace=True)
start_index = df_Limestone.columns.get_loc('DATE')
end_index = df_Limestone.columns.get_loc('Limestone Pile Qty')
selected_columns = df_Limestone.columns[start_index:end_index + 1]
df_selected_Limestone = df_Limestone[selected_columns]

df_selected_Limestone['DATE'] = pd.to_datetime(df_selected_Limestone['DATE'],errors='coerce')
cutoff_date = pd.to_datetime('2024-12-31')
df_filtered_Limestone = df_selected_Limestone[
    (df_selected_Limestone['DATE'].dt.month < 12) | 
    ((df_selected_Limestone['DATE'].dt.month == 12) & (df_selected_Limestone['DATE'].dt.day <= 31))
]
for col in df_filtered_Limestone.columns:
        if col != "DATE":
            print(f"\nProcessing Column: {col}")  # Debugging print
            print("Before Cleaning:\n", df_filtered_Limestone[col].head())  # Show raw values before cleaning
            df_filtered_Limestone[col] = clean_numeric_column(df_filtered_Limestone[col])
            print("After Cleaning:\n", df_filtered_Limestone[col].head())
            # df[col] = df[col].apply(lambda x: round(float(x), 6) if pd.notnull(x) else np.nan)

print(df_filtered_Limestone.dtypes)

df_filtered_Limestone.to_csv('testing2.csv',index=False)
```

```python
def clean_sheet_data(sheet_df):
    try:
        """Process individual sheet data into long format."""
        # Set column headers using the first valid row
        sheet_df.dropna(axis=0, how='all', inplace=True)
        
        # Remove metadata rows and empty first column
        df_cleaned = sheet_df.iloc[1:].reset_index(drop=True)
        df_cleaned = df_cleaned[0 + 1:] 
        
        # Calculate number of 5-column blocks dynamically
        df_cleaned.dropna(how='all', axis=1, inplace=True)  # Drop empty columns
        df_cleaned.dropna(how='all', inplace=True)
        num_columns = len(df_cleaned.columns)
        print(num_columns)
        num_blocks = num_columns // 5

        print(num_blocks)
        
        # Process each block
        cleaned_blocks = []
        for block_idx in range(num_blocks):
            start_col = block_idx * 5
            end_col = start_col + 5
            block = df_cleaned.iloc[:, start_col:end_col].copy()
            block.columns = ['SamplingTime', 'SO3', 'K2O', 'Na2O', 'Cl%']
            cleaned_blocks.append(block)
        
        # Combine blocks and clean
        combined = pd.concat(cleaned_blocks, ignore_index=True)
        combined = combined.dropna(subset=['SamplingTime'])
        print(combined.head())
        return combined
    except Exception as e:
        print(e)

    
```

```python
import os
import re
import pandas as pd
from datetime import datetime, timedelta

def format_hour(hour_range, base_date):
    """Format hour range with correct date handling"""
    start_hour = int(hour_range.split(" - ")[0])
    if start_hour == 23:
        date = base_date - timedelta(days=1)
    else:
        date = base_date
    return date.strftime("%d/%m/%Y") + f" {start_hour:02d}:00:00"

def extract_date(file_name):
    # print(file_name)
    match = re.search(r"(\d{2})\.(\d{2})\.(\d{3,4})", file_name)  # Match dates like 05.02.204 or 05.02.2024
    # print(match)
    if match:
        day, month, year = match.groups()
        
        # Fix incomplete year (assuming it's always 2000+)
        if year in ["204", "224"]:  # Handle special cases
            year =  "2024"
        elif len(year) == 3:
            year = "2" + year  # Convert "204" → "2024"
        elif len(year) == 4 and year.startswith("22"):
            year =  "2024"  # Fix incorrect "2224" or "2204" → "2024"
        return f"{day}.{month}.{year}"

    return "No Date Found"


def process_excel_file(file_path):
    try:
        """Process individual Excel file"""
        # Extract date from filename
        file_name = os.path.basename(file_path)
        print(file_name)
        # match = re.search(r"\b(\d{2})\.(\d{2})\.(\d{3,4})\b", file_name)
        df_cy = pd.read_excel(file_path)
        date_pattern = r"[A-Za-z]+, [A-Za-z]+ \d{1,2}, \d{4}"
        # print(df_cy.head(20))

        # Iterate through the first column to find the date
        date_range = None
        for row in df_cy.itertuples(index=False):  # Loop row by row
            for cell in row:  # Loop through each column in the row
                if isinstance(cell, str) and re.search(date_pattern, cell):  # Check if it matches the date format
                    date_range = cell.strip()  # Clean any spaces
                break
            if date_range:
                break
        print(date_range)
        # Regular expression to extract dates
        date_pattern2 = r"[A-Za-z]+, ([A-Za-z]+ \d{1,2}, \d{4})"

        # Find all matches
        matches = re.findall(date_pattern2, date_range)

        if len(matches) >= 2:
            second_date_str = matches[1]  # Extract second date
            # Convert to datetime object
            second_date = datetime.strptime(second_date_str, "%B %d, %Y")
            # Format into desired format
            formatted_date = second_date.strftime("%d.%m.%Y")
            print("Formatted Second Date:", formatted_date)
        else:
            print("No second date found")
        base_date = datetime.strptime(formatted_date, "%d.%m.%Y")
        data_start_row_index =df_cy[df_cy.iloc[:, 1].notna()].index[0] - 1  # Adjust this logic as needed
        df_filtered = df_cy[data_start_row_index + 1:].copy() 
        
        df_filtered = df_filtered.dropna(axis=1, how='all')  # Drop empty columns
        df_filtered = df_filtered.dropna(axis=0, how='all')  
        header_row_index = df_filtered[df_filtered.apply(lambda x: x.astype(str).str.contains('degC').any(), axis=1)].index[0]

        df_filtered = df_filtered.iloc[1:]  # Remove the first row
        # df_filtered.columns = df_filtered.iloc[0]  # Set first row as column names
        df_filtered.columns = df_filtered.iloc[0].values  # Set first row as column names
        df_filtered = df_filtered[1:].reset_index(drop=True)  # Drop the first row and reset index
        df_filtered = df_filtered.drop(df_filtered.columns[1], axis=1).iloc[:24]

        hour_values = ["23 - 00"] + [f"{h:02d} - {h+1:02d}" for h in range(0, 23)]
        df_filtered.insert(0, "Hour", hour_values)

        df_filtered["Hour"] = df_filtered["Hour"].apply(lambda x: format_hour(x, base_date))

        df_filtered.columns = ["Hour"] + [f"CYCLONE {i} {col} AVG" for i, col in enumerate(df_filtered.columns[1:], start=1)]

        # df_filtered.columns = new_column_names
        # print(df_filtered.columns)
        
        # df_filtered.to_csv('testingcy.csv',index=False)
        # print(df_filtered)
        return df_filtered
    except Exception as e:
        print(e)

def process_folder(main_folder):
    """Process all Excel files in folder structure"""
    all_data = []
    
    for root, dirs, files in os.walk(main_folder):
        for file in files:
            if file.endswith('.xlsx'):
                file_path = os.path.join(root, file)
                print(f"Processing {file_path}")
                
                try:
                    df = process_excel_file(file_path)
                    if df is not None:
                        all_data.append(df)
                    # print(df.head(20))
                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")
               
    # Combine all data
    if all_data:
        final_df = pd.concat(all_data, ignore_index=True)
        # final_df = final_df.iloc[:, :16]
        # final_df.to_csv(output_csv, index=False)
        return final_df
        print(f"Combined data saved to {output_csv}")
    else:
        print("No valid data found to process")

# Usage
main_folder = "/Users/sushantshinde/workspaces/trvi/AI data for Cyclones jan.2024 to jan.2025"
output_csv = "combined_cyclone_data.csv"
CYCLONE_df  = process_folder(main_folder)
CYCLONE_df.to_csv(output_csv,index=False)
```

```python
def format_hour(hour_range, base_date):
    """Format hour range with correct date handling"""
    start_hour = int(hour_range.split(" - ")[0])
    if start_hour == 23:
        date = base_date - timedelta(days=1)
    else:
        date = base_date
    return date.strftime("%d/%m/%Y") + f" {start_hour:02d}:00:00"

def extract_date(file_name):
    # print(file_name)
    match = re.search(r"(\d{2})\.(\d{2})\.(\d{3,4})", file_name)  # Match dates like 05.02.204 or 05.02.2024
    # print(match)
    if match:
        day, month, year = match.groups()
        
        # Fix incomplete year (assuming it's always 2000+)
        if year in ["204", "224"]:  # Handle special cases
            year =  "2024"
        elif len(year) == 3:
            year = "2" + year  # Convert "204" → "2024"
        elif len(year) == 4 and year.startswith("22"):
            year =  "2024"  # Fix incorrect "2224" or "2204" → "2024"
        return f"{day}.{month}.{year}"

    return "No Date Found"


def COOLER_procee_file(file_path):
    try:
        """Process individual Excel file"""
        # Extract date from filename
        file_name = os.path.basename(file_path)
        print(file_name)
        # match = re.search(r"\b(\d{2})\.(\d{2})\.(\d{3,4})\b", file_name)
        df_cy = pd.read_excel(file_path)
        date_pattern = r"[A-Za-z]+, [A-Za-z]+ \d{1,2}, \d{4}"

        # Iterate through the first column to find the date
        date_range = None
        for row in df_cy.itertuples(index=False):  # Loop row by row
            for cell in row:  # Loop through each column in the row
                if isinstance(cell, str) and re.search(date_pattern, cell):  # Check if it matches the date format
                    date_range = cell.strip()  # Clean any spaces
                break
            if date_range:
                break
        print(date_range)
        # Regular expression to extract dates
        date_pattern2 = r"[A-Za-z]+, ([A-Za-z]+ \d{1,2}, \d{4})"

        # Find all matches
        matches = re.findall(date_pattern2, date_range)

        if len(matches) >= 2:
            second_date_str = matches[1]  # Extract second date
            # Convert to datetime object
            second_date = datetime.strptime(second_date_str, "%B %d, %Y")
            # Format into desired format
            formatted_date = second_date.strftime("%d.%m.%Y")
            print("Formatted Second Date:", formatted_date)
        else:
            print("No second date found")
        base_date = datetime.strptime(formatted_date, "%d.%m.%Y")
        data_start_row_index =df_cy[df_cy.iloc[:, 1].notna()].index[0] - 1  # Adjust this logic as needed
        df_filtered = df_cy[data_start_row_index + 1:].copy() 
        df_filtered = df_filtered.dropna(axis=1, how='all')  # Drop empty columns
        df_filtered = df_filtered.dropna(axis=0, how='all')  
        header_row_index = df_filtered[df_filtered.apply(lambda x: x.astype(str).str.contains('rpm').any(), axis=1)].index[0]

        df_filtered = df_filtered.iloc[1:]  # Remove the first row
        df_filtered.columns = df_filtered.iloc[0]  # Set first row as column names
        df_filtered = df_filtered[1:].reset_index(drop=True)  # Drop the first row and reset index
        df_filtered = df_filtered.drop(df_filtered.columns[1], axis=1).iloc[:24]

        hour_values = ["23 - 00"] + [f"{h:02d} - {h+1:02d}" for h in range(0, 23)]
        df_filtered.insert(0, "Hour", hour_values)

        df_filtered["Hour"] = df_filtered["Hour"].apply(lambda x: format_hour(x, base_date))
        # df_filtered.columns = ["CYCLONE " + col for col in df_filtered.columns]

        new_column_names = ["Hour"] + [f"COOLER {i} {col} AVG" for i, col in enumerate(df_filtered.columns[1:], start=1)]

        df_filtered.columns = new_column_names
        
        # df_filtered.to_csv('testingcy.csv',index=False)
        # print(df_filtered)
        return df_filtered
    except Exception as e:
        print(e)
    

def process_folder_COOLER(main_folder):
    """Process all Excel files in folder structure"""
    all_data = []
    
    for root, dirs, files in os.walk(main_folder):
        for file in files:
            if file.endswith('.xlsx'):
                file_path = os.path.join(root, file)
                print(f"Processing {file_path}")
                
                try:
                    df = COOLER_procee_file(file_path)
                    if df is not None:
                        all_data.append(df)
                except Exception as e:
                    print(f"Error processing {file_path}: {str(e)}")
            
    # Combine all data
    if all_data:
        final_df = pd.concat(all_data, ignore_index=True)
        return final_df
        # final_df.to_csv(output_csv, index=False)
        print(f"Combined data saved to {output_csv}")
    else:
        print("No valid data found to process")

# Usage
main_folder = "/Users/sushantshinde/workspaces/trvi/AI Data For Kiln,Cooler jan.2024 to jan.2025"

COOLER_df = process_folder_COOLER(main_folder)
# COOLER_df.to_csv(output_csv, index=False)
```
